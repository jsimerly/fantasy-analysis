```python
from pulp import *
import pandas as pd
import pandas as pd
from sqlalchemy.orm import sessionmaker
from sqlalchemy import create_engine
from datetime import datetime, timedelta

from data.models import Player, KtcValue
```

```python
DB_USERNAME = os.environ['DB_USERNAME']
DB_PASSWORD = os.environ['DB_PASSWORD']
DB_HOST = os.environ['DB_HOST']
DB_PORT = os.environ['DB_PORT']
DB_NAME = os.environ['DB_NAME']

DB_URL = f"postgresql://{DB_USERNAME}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

engine = create_engine(DB_URL, echo=False)
Session = sessionmaker(bind=engine)
session = Session()
```

```python
offset = datetime.now() - timedelta(days=2)
players = session.query(Player).join(Player.ktc_values).filter(
    Player.ktc_id.is_not(None),
    KtcValue.date >= offset
).distinct().all()
```

```python
playerData = []
for player in players:
    mostRecentVal = max(player.ktc_values, key=lambda x: x.date)
    playerData.append({
        'name': player.display_name,
        'pos': player.position,
        'value': mostRecentVal.value
    })

df = pd.DataFrame(playerData)
```
